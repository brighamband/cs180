## Reading/Writing Data

```python
import requests
dta = requests.get("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data").text
headers = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)', 'class']
```

```python
dta
```

```python
rows = []
for row in dta.splitlines():
    if len(row) == 0:
      continue
    rows.append(row.split(","))

```

```python
rows
```

```python
iris = pd.DataFrame(rows, columns=headers)
iris["sepal length (cm)"] = iris["sepal length (cm)"].astype(float)
iris["sepal width (cm)"] = iris["sepal width (cm)"].astype(float)
iris["petal length (cm)"] = iris["petal length (cm)"].astype(float)
iris["petal width (cm)"] = iris["petal width (cm)"].astype(float)
```

```python
iris.head(10)
```

```python
iris.tail(10)
```

```python
iris
```

```python
# write csv
iris.to_csv("iris.csv", index=False)
```

```python
headers = [
  "age",
  "workclass",
  "fnlwgt",
  "education",
  "education-num",
  "marital-status",
  "occupation",
  "relationship",
  "race",
  "sex",
  "capital-gain",
  "capital-loss",
  "hours-per-week",
  "native-country",
  "income"
]
```

```python
adult = pd.read_csv("adult.data", header=None)
adult.columns=headers
```

```python
adult
```

```python
# adult example
```

```python
adult[["age","hours-per-week"]].groupby("age").mean()
```

```python
adult[["age","hours-per-week"]].groupby("age").mean().sort_values("hours-per-week")
```

```python
adult[["age", "sex","hours-per-week"]].groupby(["age","sex"]).mean()
```

### Cross-tabs

```python
pd.crosstab(
    index=adult["sex"],
    columns=adult['race']
)
```

### Pivot tables

```python
pd.pivot_table(
    data= adult, 
    values='hours-per-week', 
    index=["sex"],
    columns=['race', 'education'], 
    aggfunc=np.mean
)
```

## Joining

image.png

```python
iris["class"].value_counts()
```

```python
new_data = {
    "class": ["Iris-virginica", "Iris-setosa", "Iris-metallica"],
    "cost": [12.50, 9.25, 110.0]
}
cost_data = pd.DataFrame(new_data)
```

```python
cost_data
```

```python
costs = pd.merge(left=iris, right=cost_data, on="class", how="inner")
```

```python
costs.head()
```

```python
costs[costs["class"]=="Iris-versicolor"]
```
