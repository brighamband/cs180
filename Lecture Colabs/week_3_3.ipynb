#### Correlation Coefficents

Screen Shot 2021-09-17 at 11.29.44 AM.png

- A measure of linear correlation between two sets of data
- It is the ratio between the covariance of two variables and the product of their standard deviations
- it is essentially a normalised measurement of the covariance, such that the result always has a value between 1 and 1
- As with covariance itself, the measure can only reflect a linear correlation of variables, and ignores many other types of relationship or correlation. 

image.png

```python
np.corrcoef(x, y)[0][1]
```

## Box Plots

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import requests
dta = requests.get("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data").text
headers = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)', 'class']
rows = []
for row in dta.splitlines():
    if len(row) == 0:
      continue
    rows.append(row.split(","))
iris = pd.DataFrame(rows, columns=headers)
iris["sepal length (cm)"] = iris["sepal length (cm)"].astype(float)
iris["sepal width (cm)"] = iris["sepal width (cm)"].astype(float)
iris["petal length (cm)"] = iris["petal length (cm)"].astype(float)
iris["petal width (cm)"] = iris["petal width (cm)"].astype(float)
```

```python
iris.head()
```

```python
iris.boxplot()
```

## Scatter Plot Matrix

```python

pd.plotting.scatter_matrix(iris, alpha=1.0, figsize=(12, 12))
```

# Exploratory Data Analysis (EDA) - Categorical Data:

```python
headers = [
  "age",
  "workclass",
  "fnlwgt",
  "education",
  "education-num",
  "marital-status",
  "occupation",
  "relationship",
  "race",
  "sex",
  "capital-gain",
  "capital-loss",
  "hours-per-week",
  "native-country",
  "income"
]
adult = pd.read_csv("adult.data", header=None)
adult.columns=headers
```

```python
adult = adult.sample(1000)
```

```python
adult.head()
```

## Univariate Plots and Statistics


```python
cnts = adult["marital-status"].value_counts()
```

```python
cnts
```

```python
f, ax = plt.subplots(figsize=(14, 6))
sns.barplot(cnts.index,cnts)
```

### Bivariate Plots and Statistics

### Heat maps and conditional probabilities

```python
# Heat maps
heat = pd.crosstab(
    index=adult["education"],
    columns=adult['income'],
    normalize = "index"
)
```

```python
heat
```

```python
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(heat, annot=True, linewidths=.5, ax=ax)
```
