#Exercise 4

```python
import seaborn as sns
sns.regplot(x="sugars", y="rating", data=df)
```

```python
corr_matrix = np.corrcoef(df['sugars'], df['rating'])
corr = corr_matrix[0,1]
r_squared = corr**2
r_squared
```

###Simply by viewing the graph, but also by its r^2 value of `0.577...`, we learn that there is a fairly strong negative relationship between sugar and ratings.  As the amount of sugars in a sugar decreases, its rating generally decreases.

#Exercise 5

`carbs` - weak to no correlation to rating

`sugars` - strong, negative correlation to rating (cereals with less sugars generally have higher ratings)

`calories` - strong, negative correlation to rating (cereals with fewer calories generally have higher ratings)

`protein` - strong, positive correlation to rating (cereals with more protein generally have higher ratings)

`fat` - strong, negative correlation to rating (cereals with less fat generally have higher ratings)


###I would say that `sugars` has the strongest relationship with rating. While `calories`, `protein`, and `fat` all have strong correlations/relationships to rating, it appears `sugar` has the strongest relationship because with the other variables, their distributions are more concentrated on the median, therefore the differences in those variables are less in magnitude, affecting the rating less.  For sugars, however, they have a wide distribution, ranging from 0 clear up.  After all, 6 of the 7 top rated cereals have a `sugars` value of 0.
